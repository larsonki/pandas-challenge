```python
# Dependencies
import pandas as pd
import numpy as np

# Upload csv
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(school_data)
student_df = pd.read_csv(student_data)

# Confirm which column(s) to use to combine data into a dataset.
#school_df
#student_df

# Combine the data into a single dataset using shared column "school_name" & confirm object types
school_df_complete = pd.merge(student_df, school_df, how="left", on="school_name")
#school_df_complete.dtypes
school_df_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate total number of schools
total_schools = school_df["school_name"].count()
print(f"Total Schools: {total_schools}")
# Calculate total number of students
total_students = student_df["student_name"].count()
print(f"Total Students: {total_students}")
# Calculate total budget
total_budget = school_df["budget"].sum()
print(f"Total Budget: {total_budget}")
```

```python
# Calculate the average math score
avg_math_score = school_df_complete["math_score"].mean()
print(f"Average Math Score: {avg_math_score}")
# Calculate the average reading score
avg_reading_score = school_df_complete["reading_score"].mean()
print(f"Average Reading Score: {avg_reading_score}")
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
## Start by isolating the df to only reflect the rows of data in "math_score" that have values >69
stu_pass_math = school_df_complete.loc[school_df_complete["math_score"] > 69, :]
## Then count the number of scores in the new df
num_stu_pass_math = stu_pass_math["math_score"].count()
## Then calculate the percentage based off the total number of students
per_pass_math = num_stu_pass_math/total_students * 100
print(f"Percentage of Students Passing Math: {per_pass_math}")

# Calculate the percentage of students with a passing reading score (70 or greater)
## Apply same logic as above
stu_pass_read = school_df_complete.loc[school_df_complete["reading_score"] > 69, :]
num_stu_pass_read = stu_pass_read["reading_score"].count()
per_pass_read = num_stu_pass_read/total_students * 100
print(f"Percentage of Students Passing Reading: {per_pass_read}")

# Calculate the percentage of students who passed math and reading (% Overall Passing)
## Apply same logic as above, except filter using & to capture only students passing in both subjects
overall_pass = school_df_complete.loc[(school_df_complete["math_score"] > 69) & \
                                      (school_df_complete["reading_score"] > 69), :]
num_overall_pass = overall_pass["student_name"].count()
per_overall_pass = num_overall_pass/total_students * 100
print(f"Percentage of Students Passing Overall: {per_overall_pass}")
```

```python
# Create a dataframe to hold the above results
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math_score],
    "Average Reading Score": [avg_reading_score],
    "% Passing Math": [per_pass_math],
    "% Passing Reading": [per_pass_read],
    "% Overall Passing": [per_overall_pass]
})
district_summary
# Optional: give the displayed data cleaner formatting
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Rename the columns from the original merged dataframe to to make them easier to understand and closer to what final
# columns should be
renamed_df = school_df_complete.rename(columns={"student_name": "Total Students", "school_name": "School Name",\
                                                "reading_score": "Average Reading Score", "math_score": \
                                                "Average Math Score", "type": "School Type", "budget": \
                                                "Total School Budget"})

# Group data together based off "School Name" and "School Type""
grouped_school = renamed_df.groupby(["School Name", "School Type", "Total School Budget"])

# Calculate "Total Students", "Average Math Score", "Average Reading Score"
total_students_by_school = grouped_school[["Total Students"]].count().reset_index()
total_students_by_school["Per Student Budget"] = total_students_by_school["Total School Budget"]/\
                                                    total_students_by_school["Total Students"]

avg_math_score_by_school = grouped_school[["Average Math Score"]].mean().reset_index()
avg_math_score_by_school = avg_math_score_by_school.drop(["School Name", "School Type",
                                                          "Total School Budget"], axis=1)
avg_reading_score_by_school = grouped_school[["Average Reading Score"]].mean().reset_index()
avg_reading_score_by_school = avg_reading_score_by_school.drop(["School Name", "School Type",
                                                                "Total School Budget"], axis=1)
```

```python
# Calculate % passing math by first filtering out passing scores, grouping them by school to count total, converting
# to a list and appending that list to a previously used DataFrame to complete calculation. Drop the superfluous
# column
filtered_math = renamed_df.loc[renamed_df["Average Math Score"]>69,:]
grouped_math = filtered_math.groupby(["School Name"])
num_pass_math_by_school = pd.DataFrame(grouped_math[["Average Math Score"]].count())
col_list = list(num_pass_math_by_school["Average Math Score"])
total_students_by_school["# Students Passing Math"] = col_list
total_students_by_school["% Passing Math"] = total_students_by_school["# Students Passing Math"]/\
                                                total_students_by_school["Total Students"] *100
total_students_by_school = total_students_by_school.drop("# Students Passing Math", axis=1)
total_students_by_school
```

```python
# Calculate % passing reading using same logic as % passing math
filtered_reading = renamed_df.loc[renamed_df["Average Reading Score"]>69,:]
grouped_reading = filtered_reading.groupby(["School Name"])
num_pass_reading_by_school = pd.DataFrame(grouped_reading[["Average Reading Score"]].count())
col2_list = list(num_pass_reading_by_school["Average Reading Score"])
total_students_by_school["# Students Passing Reading"] = col2_list
total_students_by_school["% Passing Reading"] = total_students_by_school["# Students Passing Reading"]/\
                                                total_students_by_school["Total Students"] *100
total_students_by_school = total_students_by_school.drop("# Students Passing Reading", axis=1)
total_students_by_school
```

```python
# Calculate % overall passing using same logic as % passing math
overall = renamed_df.loc[(renamed_df["Average Math Score"] > 69) & \
                                      (renamed_df["Average Reading Score"] > 69), :]
grouped_overall = overall.groupby(["School Name"])
num_overall = pd.DataFrame(grouped_overall[["School Name"]].count())
col3_list = list(num_overall["School Name"])
total_students_by_school["# Overall"] = col3_list
total_students_by_school["% Overall Passing"] = total_students_by_school["# Overall"]/\
                                                total_students_by_school["Total Students"] *100
total_students_by_school = total_students_by_school.drop("# Overall", axis=1)
total_students_by_school
```

```python
# Create DataFrame to hold above results
school_summary = pd.concat([total_students_by_school, avg_math_score_by_school, avg_reading_score_by_school], 
                             axis=1)
school_summary = school_summary.set_index("School Name")
org_school_summary = school_summary[["School Type", "Total Students", "Total School Budget", "Per Student Budget",\
                                    "Average Math Score", "Average Reading Score", "% Passing Math",\
                                    "% Passing Reading", "% Overall Passing"]]
org_school_summary
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting



```python
# Create a DataFrame that includes the average math score for each grade level at each school
## Isolate grade levels using groupby and apply the mean
grade_nine = renamed_df.loc[renamed_df["grade"] == "9th", :]
group_nine = grade_nine.groupby(["School Name"])
mean_nine = group_nine[["Average Math Score"]].mean().reset_index()

grade_ten = renamed_df.loc[renamed_df["grade"] == "10th", :]
group_ten = grade_ten.groupby(["School Name"])
mean_ten = group_ten[["Average Math Score"]].mean().reset_index()

grade_eleven = renamed_df.loc[renamed_df["grade"] == "11th", :]
group_eleven = grade_eleven.groupby(["School Name"])
mean_eleven = group_eleven[["Average Math Score"]].mean().reset_index()

grade_twelve = renamed_df.loc[renamed_df["grade"]=="12th", :]
group_twelve = grade_twelve.groupby(["School Name"])
mean_twelve = group_twelve[["Average Math Score"]].mean().reset_index()
## Create a DataFrame importing the means from each grade.
math_by_grade = mean_nine
math_by_grade["10th"] = mean_ten["Average Math Score"]
math_by_grade["11th"] = mean_eleven["Average Math Score"]
math_by_grade["12th"] = mean_twelve["Average Math Score"]
math_by_grade = math_by_grade.rename(columns={"Average Math Score":"9th"})
## Reset index to "School Name"
math_by_grade = math_by_grade.set_index("School Name")
math_by_grade
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
# Create a DataFrame that includes the average reading score for each grade level at each school
## Isolate grade levels using groupby and apply the mean
grade_nine_r = renamed_df.loc[renamed_df["grade"] == "9th", :]
group_nine_r = grade_nine_r.groupby(["School Name"])
mean_nine_r = group_nine_r[["Average Reading Score"]].mean().reset_index()

grade_ten_r = renamed_df.loc[renamed_df["grade"] == "10th", :]
group_ten_r = grade_ten_r.groupby(["School Name"])
mean_ten_r = group_ten_r[["Average Reading Score"]].mean().reset_index()

grade_eleven_r = renamed_df.loc[renamed_df["grade"] == "11th", :]
group_eleven_r = grade_eleven_r.groupby(["School Name"])
mean_eleven_r = group_eleven_r[["Average Reading Score"]].mean().reset_index()

grade_twelve_r = renamed_df.loc[renamed_df["grade"]=="12th", :]
group_twelve_r = grade_twelve_r.groupby(["School Name"])
mean_twelve_r = group_twelve_r[["Average Reading Score"]].mean().reset_index()
## Create a DataFrame importing the means from each grade.
reading_by_grade = mean_nine_r
reading_by_grade["10th"] = mean_ten_r["Average Reading Score"]
reading_by_grade["11th"] = mean_eleven_r["Average Reading Score"]
reading_by_grade["12th"] = mean_twelve_r["Average Reading Score"]
reading_by_grade = reading_by_grade.rename(columns={"Average Reading Score":"9th"})
## Reset index to "School Name"
reading_by_grade = reading_by_grade.set_index("School Name")
reading_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Drop unnecessary columns from original school summary DataFrame, create bins & labels, perform cut function
# and groupby "Per Student Budget", calculate mean within the group
spending_summary = org_school_summary.drop(["School Type", "Total Students", "Per Student Budget",\
                                            "Total School Budget"], axis=1)
bins = [0, 584, 629, 644, 681]
group_names = ["<$585", "$586-630", "$631-645", "$646-680"]
spending_summary["Spending Ranges (Per Student)"] = pd.cut(org_school_summary["Per Student Budget"],\
                                                             bins, labels=group_names, include_lowest=True)
spending_group = spending_summary.groupby("Spending Ranges (Per Student)")
spending_group.mean()
```
